### Laplace equation
It's a special case of Poisson's equation. Here, $f(x,y)=0$. We can take the equations,
$$\frac{\partial^2 U}{\partial x^2} + \frac{\partial^2 U}{\partial y^2} = 0;\quad 0\leq x\leq 1,\quad 0\leq y \leq 1$$
$$ U(x,0)=U(x,1)=x,\quad 0\leq x\leq 1 $$
$$ U(0,y)=U(1,y)=y,\quad 0\leq y\leq 1 $$
