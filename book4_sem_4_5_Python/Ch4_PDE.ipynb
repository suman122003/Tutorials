We would have linear PDE of the form,
$$A\frac{\partial^2 U}{\partial x^2} + B\frac{\partial^2 U}{\partial x \partial y} + C\frac{\partial^2 U}{\partial y^2} = g(x,y)$$
Here $A$, $B$ and $C$ are constants.

Categories of PDE:
1. **Parabolic**: $B^2 -4AC = 0$
2. **Elliptic**: $B^2 -4AC < 0$
3. **Hyperbolic**: $B^2 -4AC > 0$
