### `countplot`
**Difference between countplot and barplot**: Countplot plots the count of the number of records by category. Barplots plots a value or metric for each category (by default barplot plots the mean of a variable by category).

```python
#display(stats1.head(), stats3.head())
sns.set_style('whitegrid')
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(x='Division', hue='Club', data=stats1, 
            palette='winter_r', saturation=0.9, alpha=0.7)
plt.xticks(rotation=30)
plt.title('no. of seasons played by Messi')

plt.subplot(1,2,2)
sns.countplot(x='Division', hue='Club', data=stats3, 
            palette='winter_r', saturation=0.9, alpha=0.7)
plt.xticks(rotation=30)
plt.title('no. of seasons played by Ronaldo')
plt.show()
```

### `violinplot`
A violin plot plays a similar role as a boxplot (whisker plot). It shows the distribution of quantitative data across several levels of one (or more) categorial varaibles such that those distributions can be compared.

**`scale`**: 'count', 'width', 'area'

**`inner`**: None, 'box', 'point', 'stick', 'quart', 'quartile'

```python
#display(stats1.head(), stats3.head())
sns.set_style('whitegrid')
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
sns.violinplot(x='Division', y='Total Goals', data=stats1, hue='Club', 
        linewidth=2, palette='Set1', order=['La Liga','Ligue 1'], 
        saturation=0.6, split=False, scale='count', inner='box')
plt.xticks(rotation=30)
plt.title('Total club goals by Messi')

plt.subplot(1,2,2)
sns.violinplot(x='Division', y='Total Goals', data=stats3, hue='Club', 
        linewidth=2, palette='Set1', saturation=0.6, 
        split=False, scale='count', inner='box')
plt.xticks(rotation=30)
plt.title('Total club goals by Ronaldo')
plt.show()
```

### `pairplot`

Axes: **`vars`** or (**`x_vars`** and **`y_vars`**)

**`kind`** (or, `diag_kind`): 'scatter', 'kde', 'hist', 'reg'.

```python
display(stats1.head())

sns.pairplot(stats1, vars=['League Apps','Total Apps','League Goals','Total Goals'], 
            hue='Club', hue_order=['PSG','Barcelona'], 
            palette='rocket_r', diag_kind='hist', markers=['<','o'])
plt.show()
```

```python
sns.pairplot(stats1, x_vars=['League Apps','Total Apps'], 
            y_vars=['League Goals','Total Goals'], hue='Club', kind='reg')
plt.show()
```

### `stripplot`
It is basically a `scatterplot` that differentiates different categories. Strip plots are considered as a good alternative to a `boxplot` or a `violinplot`.

```python
#display(stats1.head(), stats3.head())
sns.set_style('whitegrid')
plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
sns.stripplot(x='Division', y='Total Goals', data=stats1, hue='Club', jitter=1, 
            palette='rocket', linewidth=0.5, edgecolor='blue', size=5, alpha=0.9)
plt.xticks(rotation=30)
plt.title('Total club goals by Messi')

plt.subplot(1,2,2)
sns.stripplot(x='Division', y='Total Goals', data=stats3, hue='Club', jitter=1, 
            palette='rocket', linewidth=0.5, edgecolor='blue', size=5, alpha=0.9)
plt.xticks(rotation=30)
plt.title('Total club goals by Ronaldo')
plt.show()
```

### `boxplot`
In plot: Minimum non-outlier, box, median, maximum non-outlier.

```python
#display(stats1.head(), stats3.head())
sns.set_style('whitegrid')
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(x='Division', y='League Goals', data=stats1, hue='Club', order=['La Liga','Ligue 1'], 
            showmeans=True, meanprops={'marker':'+','markeredgecolor':'g'}, linewidth=2, orient='v')
plt.xticks(rotation=30)
plt.title('League goals by Messi')

plt.subplot(1,2,2)
sns.boxplot(x='Division', y='League Goals', data=stats3, hue='Club', 
            showmeans=True, meanprops={'marker':'+','markeredgecolor':'g'}, linewidth=2, orient='v')
plt.xticks(rotation=30)
plt.title('League goals by Ronaldo')

plt.show()
```
