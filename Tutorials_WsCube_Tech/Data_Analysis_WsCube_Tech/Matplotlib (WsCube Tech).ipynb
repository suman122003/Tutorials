## Data visualization
It's a quick, easy way to convey concepts in a universal manner. Data visualization is the graphical representation of information and data. The visual elements are charts, graphs and maps. Visualization provides an accessible way to see and understand trends, outliers and patterns in data.

## Data visualization in python
The python libraries for plotting are -
1. **`matplotlib`**: low level, provides lots of freedom
2. **`pandas visualization`**: easy to use interface, built on `matplotlib`
3. **`seaborn`**: high level interface, great default styles
4. **`ggplot`**: based on R's ggplot2, uses grammar of graphics
5. **`plotly`**: can create interactive plots

## Matplotlib

```python
import matplotlib.pyplot as plt
#(or, from matplotlib import pyplot as plt)
import numpy as np
```

### `bar` plot

```python
x1 = range(0,20,2)
y1 = np.random.randn(10)
c1 = ['violet','b','g','y','r']*2

plt.bar(x1, y1, color=c1)
plt.show()
```

```python
language = ['python', 'C', 'C++', 'Java']
popularity = [85,70,60,82]
level = [40,10,20,30]

pr = np.arange(len(language))
w = 0.2
pr1 = pr + w  # or [j+w for j in pr]

plt.bar(pr, popularity, width=w, color=['g'], align='center', 
    edgecolor='r', linewidth=2, linestyle=':', alpha=0.6, label='popularity')
# alpha = 0 to 1
plt.bar(pr1, level, width=w, color='m', label='language level')
plt.xticks(pr+w/2, language, rotation=30)
plt.xlabel('language')
plt.ylabel('relative values')
plt.title('relative comparisons', fontsize=15)
plt.legend()
plt.show()
```

```python
language = ['python', 'C', 'C++', 'Java']
popularity = [85,70,60,82]

plt.barh(language, popularity, color=['g'], align='center', 
    alpha=0.7, label='popularity')
plt.title('relative comparisons (horizontal)', fontsize=15)
plt.legend()
plt.show()
```

### `scatter` plot

```python
size1 = [100,300,180,260,400]*2
plt.scatter(x1, y1, color=c1, s=size1, alpha=0.8, marker='*', 
            edgecolor='black', linewidth=2)
plt.show()
```

```python
x3 = np.arange(0,20,2)
y3 = np.linspace(0,10,10) % 5
y3p = np.array([5,6,9,3,8,2,0,7,3,5])
c3 = [50,15,42,10,35]*2

plt.scatter(x3, y3, c='r', s=50, label='% 5')
plt.scatter(x3, y3p, c=c3, s=50, cmap='viridis', alpha=0.6)
cb = plt.colorbar()
cb.set_label('Colorbar')
plt.legend()
plt.show()
```

### Histogram plot (hist)

```python
x4 = [11, 22, 54, 59, 50, 46, 22, 39, 41, 49, 34, 19, 41, 17, 47, 48, 51,
       10, 31, 57, 55, 54, 52, 31, 17, 27, 38, 12, 40, 25, 28, 36, 29, 34,
       40, 25, 44, 24, 26, 33, 28, 41, 37, 56, 25, 18, 33, 21, 26, 37]
list1 = [10,20,30,40,50,60]

plt.hist(x4, bins=list1, color='g',range=(0,100), edgecolor='r', cumulative=-1, 
    bottom=30, align='mid', histtype='bar', orientation='vertical', rwidth=8, log=True)
# range is not working (1:51:40)
plt.axvline(35, color='m', label='midline')
plt.title('histogram plot')
plt.legend()
plt.show()
```

### pie plot

```python
display(language, popularity, level)

c2 = ['y', 'indigo', 'm', 'g']
ex = [0,0.35,0,0]

plt.pie(level, labels=language, labeldistance=1.2, rotatelabels=False, 
        textprops={'fontsize': 12}, explode=ex, colors=c2, shadow=True, 
        wedgeprops={'linewidth':4.5, 'width':0.8, 'edgecolor':'grey'}, 
        radius=1, center=(3,5), startangle=-180, counterclock=True, autopct='%0.2f%%')
plt.title('language levels (pie chart)')
plt.legend()
plt.show()
```

### `stem` plot
**Linefmt**:
1. `'-'` --- solid line
2. `'--'` --- dashed line
3. `'-.'` --- dash-dot line
4. `':'` --- dotted line
**`markerfmt`**: (marker color) (marker symbol). Example: 'r o', 'g *', 'y+'.

```python
x5 = np.arange(2,14,2)
y5 = [5,2,3,7,2,6]

plt.stem(x5, y5, linefmt=':', markerfmt='r +', use_line_collection=False,
        bottom=np.pi, basefmt='g', label='reference', orientation='vertical')
plt.title('stem plot')
plt.legend()
plt.show()
```

### `boxplot`
The line inside the box (between the min and max of the data given) is called *median line*. 3 lines 1st, 2nd (median) and 3rd quartile line are present at the boundary and inside the box. The min and max of the data are indicated by *whisker lines*. Value outside range is represented by a point (*outlier*) outside box.

```python
x6 = np.arange(2,9)*10
x6 = np.append(x6, [140,10])
x6a = np.arange(4,12)*10
display(x6, x6a)

plt.boxplot([x6, x6a], notch=False, vert=False, widths=0.15, labels=['data1', 'data2'], 
    patch_artist=False, showmeans=True, whis=1.6, sym='b', boxprops={'color':'g'}, 
    capprops={'color':'r'}, whiskerprops={'color':'b'}, 
    flierprops={'markeredgecolor':'o'})  # flierprops ?
plt.show()
```

### `stackplot` (area plot)

```python
x7 = np.arange(1,11)
y7a = np.array([8,6,4,8,6,8,3,7,5,7])
y7b = np.array([4,8,3,6,2,1,2,3,4,2])
y7c = np.array([4,6,8,3,5,6,5,9,8,7])
display(x7, y7a, y7b, y7c)

lbls = ['area1', 'area2', 'area3']
c4 = ['y','g','b']

plt.stackplot(x7, y7a, y7b, y7c, labels=lbls, colors=c4, baseline='zero')
plt.xlabel('x')
plt.ylabel('areas')
plt.legend(loc=8)
plt.title('stackplot')
plt.show()
```

### `step` plot

```python
x8 = np.array([1,2,3,4,5])
y8 = np.array([5,8,3,6,4])

plt.step(x8, y8, color='g', marker='o', mfc='r', label='random data')
plt.legend()
plt.title('step plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

### `fill_between` plot

```python
x9 = np.linspace(0, 3*np.pi, 50)
y9 = np.sin(x9)

plt.plot(x9, y9, color='r')
plt.fill_between(x9, y9, where= (x9>np.pi/2) & (x9<5*np.pi/2), 
                color='y', alpha=0.5)
plt.title('fill_between plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### `savefig`
Files will be in .png form by default.

**`dpi`**: dots per inches

```python
x10a = np.arange(3,15,2)
y10a = np.array([4,8,3,9,4,2])
x10b = np.array([1,5,3,9,4])*1e4
x10c = ['a', 'b', 'c', 'd', 'e']
y10c = [4,8,3,1,4]

plt.subplot(2,2,1)
plt.scatter(x10a, y10a)

plt.subplot(2,2,2)
plt.pie(x10b, radius=1.2)

plt.subplot(2,2,3)
plt.bar(x10c, y10c)

plt.savefig('graph 1 - matplotlib by wscube tech.pdf', 
        dpi=1000, facecolor='grey', transparent=False)
plt.show()
```

### Axes in graphs

```python
x11 = np.linspace(0,15,100)
y11 = np.cos(x11)

plt.plot(x11, y11)

xtk = np.arange(0,7)*np.pi/2
lbls1 = [0,'$\pi$/2','$\pi$','3$\pi$/2','2$\pi$','5$\pi$/2','3$\pi$']
ytk = np.array([0,-1,-3**0.5/2,-1/2**0.5,-0.5,1,3**0.5/2,1/2**0.5,0.5])

plt.xticks(xtk, labels=lbls1)
plt.yticks(ytk)

#plt.xlim(0, 3*np.pi+0.5)
#plt.ylim(-1.1,1.1)
plt.axis([0, 3*np.pi+0.5, -1.1, 1.1])
plt.grid()
plt.show()
```

### Texts in graphs
1. **`text`**: add text at an arbitary location of axes
2. **`annotate`**: add an annotation with an optional arrow at an arbitary location of axes
3. **`xlabel`**: add a label to the axes's x-axis
4. **`ylabel`**: add a label to the axes's y-axis
5. **`title`**: add a title to the axes

**`loc`**: 0 to 10 for different positions

```python
plt.plot(x11, y11)
plt.title('sinusoidal graph', fontsize=15, color='r')
plt.xlabel('x', color='b')
plt.ylabel('y', color='b')

plt.text(2,0.8,'cosine graph', fontsize=12, 
        style='italic', bbox={'facecolor':'y'})
plt.annotate('random plot', xy=(11,-0.8), xytext=(11,-0.2), 
        arrowprops={'facecolor':'r', 'shrink':50})
plt.legend('cos(x)', loc=0, 
        facecolor='g', framealpha=0.2, edgecolor='b', shadow=False)
plt.show()
```
