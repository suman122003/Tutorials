# `scipy.fft`

Documentation link: https://docs.scipy.org/doc/scipy/reference/fft.html#module-scipy.fft

```python
import scipy.fft as sfft
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# plt.style.use(['classic', 'dark_background'])
```

## `scipy.fft.fft`
If x is a 1d array, then the fft is equivalent to:

`y[k] = np.sum(x * np.exp(-2j * np.pi * k * np.arange(n)/n))`

```python
t = np.arange(256)
fnt = sfft.fft(np.sin(t))
fq = sfft.fftfreq(t.size)
sp = sfft.fftshift(fnt)
freq = sfft.fftshift(fq)
# display('fft', fnt, 'fftfreq', fq)

plt.plot(freq, sp.real, freq, sp.imag)
plt.grid()
plt.show()
```

## `scipy.fft.ifft`

If x is a 1-D array, then the ifft is equivalent to

`y[k] = np.sum(x * np.exp(2j * np.pi * k * np.arange(n)/n)) / len(x)`

```python
sfft.ifft([0,4,0,0])
```

```python
rng = np.random.default_rng()
t = np.arange(400)
n= np.zeros((400,), dtype=complex)
n[40:60] = np.exp(1j*rng.uniform(0, 2*np.pi, (20,)))
s = sfft.ifft(n)
plt.plot(t, s.real, t, s.imag, '--', lw=0.7)
plt.legend(('real part', 'imaginary part'))
plt.grid()
plt.show()
```
