```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
from scipy.linalg import eigh_tridiagonal
```

Equations to solve:
$$ \sqrt{E+V_0} \tan{\sqrt{E+V_0}} = \sqrt{-E} $$
$$ -\sqrt{E+V_0} \cot{\sqrt{E+V_0}} = \sqrt{-E} $$

```python
V0 = 36
E = np.linspace(-V0*(1-1e-5), 0, 1000)

rhs = np.sqrt(-E)
lhs1 = np.sqrt(E+V0)*np.tan(np.sqrt(E+V0))
lhs2 = -np.sqrt(E+V0)/np.tan(np.sqrt(E+V0))

plt.scatter(E, lhs1, s=1)
plt.scatter(E, lhs2, s=1)
plt.scatter(E, rhs, s=1, color='red')
plt.ylim(-10, 10)
plt.xlabel('E')
plt.grid(color='grey')
plt.show()
```

```python
def eq1(E, V0):
    return lhs1 - rhs
def eq2(E, V0):
    return lhs2 - rhs
```

```python
plt.scatter(E, lhs1-rhs, s=1)
plt.scatter(E, lhs2-rhs, s=1)
plt.scatter(E, rhs, s=1, color='red')
plt.ylim(-10, 10)
plt.xlabel('E')
plt.grid(color='grey')
plt.show()
```
