## Quick plotting

**Line plots**

```python
x1 = np.linspace(0,10,30)
y1 = np.sin(x1) + 0.1*np.random.randn(len(x1))
```

```python
plt.plot(x1,y1, color='violet', lw= 5)
plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.figure(figsize=(12,5))
plt.title('sin graph with added impurities', fontsize= 25)
plt.plot(x1,y1,'o--', ms= 12)
```

```python
x2 = np.linspace(0,10,100)
y2 = np.sin(x2)
```

```python
plt.plot(x1,y1,'o', color='red', label='graph 1')
plt.plot(x2,y2,'-', color='green', label='graph 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc= 'lower right', fontsize= 14, ncol=2)
```

x and y limits

```python
plt.plot(x1,y1,'o', color='red', label='graph 1')
plt.plot(x2,y2,'-', color='green', label='graph 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.ylim(bottom=-1.3, top=1.5)
plt.legend(loc= 'lower right', fontsize= 14, ncol=2)
```

### A more advanced API

For more complicated plots, such as

Multiple subplots on 1 figure
Conveniently adding text to plots
Creating animations
one needs to use a slightly more advanced API

**Single axis in a figure**

```python
fig, ax = plt.subplots(1,1, figsize=(10,4))
ax.plot(x1, y1, 'o--', color='r', lw=0.4, ms=3)
ax.text(0.15,0.3, 'sin like curve', fontsize=14, transform=ax.transAxes)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

**Multiple axis in a figure**

```python
fig, axes = plt.subplots(3,2, figsize=(8,9))
ax = axes[0][0]
ax.plot(x1, y1, 'o--', color='r', lw=0.5, ms=3, label='expt')
ax.text(0.3,0.2, 'sin like curve', fontsize=9)
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax = axes[1][0]
ax.plot(x2, y2, color='g', lw=1, ms=3, label='theoratical')
ax.text(0.3,0.2, 'proper sin curve', fontsize=9)
ax.set_title('ideal graph')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
