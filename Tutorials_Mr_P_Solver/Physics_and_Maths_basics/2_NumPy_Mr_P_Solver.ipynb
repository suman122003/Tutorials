**Question 4**

$$ P = A \sigma \epsilon T^4 $$ 
or, $$ P / (A \sigma \epsilon) = T^4 $$

or,$$ P_1 = P / (A \sigma \epsilon T_0^4) = \left( \frac{1}{1+e^{-kt}} \right)^4  $$

```python
kt= np.linspace(0,10,100)
P1= (1/(1 + np.exp(-kt)))**4
E= np.cumsum(P1) * (kt[1]-kt[0])
```

```python
plt.plot(kt, P1)
plt.xlabel('kt', fontsize=15)
plt.ylabel('$ P / (A \sigma \epsilon T_0^4) $', fontsize=15)
```

```python
plt.plot(kt, E)
plt.xlabel('kt', fontsize=15)
plt.ylabel(r'$ \left(\frac{k}{A \sigma \epsilon T_0^4} \right) E $', fontsize=15)
```
