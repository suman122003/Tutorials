# Numerical solution for Infintie 1D Potential Well - Shooting method

Video Link: https://youtu.be/ZSRDR5y3hwA?si=LFpWTqR1FCVmTTUL

Method:
- $ x = x + dx $
- $ \psi = \psi + \dot{\psi}dx $
- $ \dot{\psi} = \dot{\psi} + \ddot{\psi}dx $
- $ \ddot{\psi} = -k\psi (E-U) $

```python
import numpy as np
import matplotlib.pyplot as plt
```

Constants:
* Energies are in eV, lengths are in nm.
* $\hbar c = 1240eV$.
* Masses: $mc^2$ in keV.
* $kc^2 = 1.44 eV-nm$.

```python
a = 1
U = 0
m = 511e3
hbarc = 197
k = 2*m/hbarc**2

E = 0  # INPUT
dE = 0.001
psi = 0
psidot = 1
psiddot = -k*psi*E

x = 0
dx = 0.001

xp, psip = [], []
psifinal = 1

while np.abs(psifinal)>0.002:
    x = 0
    psidot = 1/np.sqrt(0.05123183508607787) # put the area value
    xp, psip = [], []
    while x<a:
        psiddot = -k*psi*(E-U)
        psidot += psiddot*dx
        psi += psidot*dx
        psip.append(psi)
        x += dx
        xp.append(x)
    psifinal = psi
    E += dE

plt.plot(xp, psip, label=f'E={E:.5f}')
plt.xlabel('x')
plt.ylabel('$\psi$')
plt.legend()
plt.grid()
plt.show()
```

```python
area = 0
for i in range(len(psip)-1):
    dA = psip[i]**2*dx
    area += dA
print(area)
```
